```python
from manim import *

config.background_color = BLACK
config.frame_width = 9
config.frame_height = 16
config.pixel_width = 1440
config.pixel_height = 2580
config.frame_rate = 60
```

```python
%%manim -v Warning --disable_caching -p Quadratics

class Quadratics(Scene):
    
    def construct(self):

    
        #Text 
        text = MathTex(r"x^2 + 10x = 39")
        self.play(Write(text))
        self.wait()
        self.play(text.animate.move_to([-2, 5, 0]))
        self.wait()
      
        #Theorem's animation:

        line_1 = Line([-2,3,0], [0, 3, 0], color=BLUE)
        self.play(Create(line_1))

        text_1 = MathTex(r"x")
        text_1.shift([-1,3.5,0])
        self.play(Write(text_1))

        line_2 = Line([-2, 3, 0], [-2, 1, 0], color = BLUE)
        self.play(Create(line_2))

        text_2 = MathTex(r"x")
        text_2.shift([-2.5, 2,0])
        self.play(Write(text_2))

        line_3 = Line([0, 3, 0], [0, 1, 0], color=BLUE)
        line_4 = Line([-2, 1, 0], [0, 1, 0], color=BLUE)

        self.play(Create(line_3), Create(line_4))

        line_5 = Line([0,3,0], [3,3,0], color = GREEN)
        text_3 = MathTex(r"5")
        text_3.shift([1.5, 3.5, 0])


        line_6 = Line([-2,1,0], [-2,-2,0], color = GREEN)
        text_4 = MathTex(r"5")
        text_4.shift([-2.5, -0.5, 0])

        self.play(Create(line_5), Write(text_3), Create(line_6), Write(text_4))
        self.wait(0.5)


        line_7 = Line([-2, -2, 0], [0, -2, 0], color=BLUE)
        line_8 = Line([0, -2, 0], [0, 1, 0], color=GREEN)
        line_9 = Line([0,1,0], [3, 1, 0], color=GREEN)
        line_10 = Line([3, 3, 0], [3,1,0], color=BLUE)
        
        self.play(Create(line_7), Create(line_8), Create(line_9), Create(line_10))
        self.wait()

        text_5= MathTex(r"x^2")
        text_5.shift([-1,2,0])

        text_6= MathTex(r"5x")
        text_6.shift([1.5, 2, 0])

        text_7= MathTex(r"5x")
        text_7.shift([-1, -0.5, 0])

        self.play(FadeIn(text_5), FadeIn(text_6), FadeIn(text_7))

        self.play(Indicate(text, color=PURPLE))

        self.play(Uncreate(text_5), Uncreate(text_6), Uncreate(text_7))

        line_11 = DashedLine([3,1,0], [3, -2, 0], color=GREEN)
        line_12 = DashedLine([0, -2, 0], [3, -2, 0], color=GREEN)

        self.play(Create(line_11), Create(line_12))

        text_8 = MathTex(r"25")

        self.play(Indicate(VGroup(line_8, line_9, line_11, line_12), color=GREEN), Write(text_8.shift([1.5, -0.5, 0])))

        square = VGroup(line_1, line_2, line_5, line_6, line_7, line_10, line_11, line_12, text_1, text_2, text_3, text_4)
        self.play(Indicate(square, color=PURPLE, rate_func=there_and_back, run_time= 2))
        self.wait()

        text_9 = MathTex(r"(x+5)^2 ="," 39 + ")
        text_9.shift([0,-3, 0])
        self.play(Write(text_9), text_8.animate.move_to([2, -3, 0]))

        text_10 = VGroup(text_9[1], text_8)
        text_11 = MathTex(r"64")
        text_12 = MathTex(r"8^2")

        self.play(Transform(text_10, text_11.shift([1, -3, 0])))
        self.wait()
        
        self.play(Transform(text_10, text_12.move_to([1, -3, 0])))

        text_13 = MathTex(r"x + 5 = 8")
        text_13.shift([0, -4, 0])

        self.play(Write(text_13))

        text_14 = MathTex(r"x = 3")
        text_14.shift([0, -5, 0])
        
        self.play(Write(text_14))
        self.wait()
        

        
      
```
